```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

## Load Image

```python
img1_jpg = cv2.imread('vdd.jpg')
img1_jpg = cv2.cvtColor(img1_jpg, cv2.COLOR_BGR2RGB)
```

# Low pass filter 

### Blur 

```python
blur = cv2.blur(img1_jpg, (13,13)) 
```

### Gaussian Blur

```python
gBlur = cv2.GaussianBlur (img1_jpg, (13,13), 0)
```

### median filter

```python
median = cv2.medianBlur(img1_jpg, 13) 
```

### Bilateral Filter 

```python
bilateral = cv2.bilateralFilter(img1_jpg, 13, 75, 75)
```

```python
titles1_jpg = ['Original Image', 'Blur', 'GaussianBlur', 'Median Filter', 'Bilateral Filter']
images1_jpg = [img_jpg, blur, gBlur, median, bilateral]
```

```python
for i in range(5):
    plt.subplot(3, 2, i+1), plt.imshow(images1_jpg[i], 'gray')
    plt.title(titles_jpg[i])
    plt.xticks([]), plt.yticks([])

fig = plt.gcf()
fig.set_size_inches(12.5, 7.5)
```
