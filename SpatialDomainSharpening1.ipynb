```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

## Load Image

```python
img1_jpg = cv2.imread('vdd.jpg')
img1_jpg = cv2.cvtColor(img1_jpg, cv2.COLOR_BGR2RGB)

img1_png = cv2.imread('vdd.png')
img1_png = cv2.cvtColor(img1_png, cv2.COLOR_BGR2RGB)

img1_tiff = cv2.imread('vdd.tiff')
img1_tiff = cv2.cvtColor(img1_tiff, cv2.COLOR_BGR2RGB)
```

# High pass filter (Sharpening) 

### Sharpening Filter 1 (Laplacian Filter)  

```python
kernel1 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
```

```python
sharpening1_jpg = cv2.filter2D(img1_jpg, -1, kernel1) 

sharpening1_png = cv2.filter2D(img1_png, -1, kernel1)

sharpening1_tiff = cv2.filter2D(img1_tiff, -1, kernel1)
```

# Sharpening filter 2

```python
kernel2 = np.array([[-1, -1, -1], [-1, 12, -1], [-1, -1, -1]])

sharpening2_jpg = cv2.filter2D(img1_jpg, -1, kernel2)

sharpening2_png = cv2.filter2D(img1_png, -1, kernel2)

sharpening2_tiff = cv2.filter2D(img1_tiff, -1 , kernel2)
```

# Sharpening Filter 3 (Add weighted Method) 

```python
sharpening3_jpg = cv2.addWeighted(img1_jpg, 4, cv2.blur(img1_jpg, (30, 30)), -4, 128)

sharpening3_png = cv2.addWeighted(img1_png, 4, cv2.blur(img1_png, (30, 30)), -4, 128)

sharpening3_tiff = cv2.addWeighted(img1_tiff, 4, cv2.blur(img1_tiff, (30, 30)), -4, 128)
```

```python
titles1_jpg = ['Original Image', 'Laplacian Filter', 'Laplacian Filter2', 'AddWeighted Filter']
images1_jpg = [img1_jpg, sharpening1_jpg, sharpening2_jpg, sharpening3_jpg]
```

```python
for i in range(4):
    plt.subplot(2, 2, i+1), plt.imshow(images1_jpg[i], 'gray')
    plt.title(titles1_jpg[i])
    plt.xticks([]), plt.yticks([])

fig = plt.gcf()
fig.set_size_inches(12.5, 7.5)
```

## Histogram for JPG File 

```python
for i in range(4):
    plt.subplot(2, 2, i+1), plt.hist(images1_jpg[i].ravel(), 256, [0, 256])
    plt.title(titles1_jpg[i])
    plt.xticks([]), plt.yticks([])

fig = plt.gcf()
fig.set_size_inches(16.5, 9.5)
```
