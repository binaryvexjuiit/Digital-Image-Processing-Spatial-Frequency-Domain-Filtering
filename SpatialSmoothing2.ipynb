```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

## Load Image

```python
img2_jpg = cv2.imread('can.jpg')
img2_jpg = cv2.cvtColor(img2_jpg, cv2.COLOR_BGR2RGB)

img2_png = cv2.imread('bea.png')
img2_png = cv2.cvtColor(img2_png, cv2.COLOR_BGR2RGB)

img2_tiff = cv2.imread('k.tiff')
img2_tiff = cv2.cvtColor(img2_tiff, cv2.COLOR_BGR2RGB)
```

# Low pass filter 

### Blur 

```python
blur_jpg = cv2.blur(img2_jpg, (13,13)) 

blur_png = cv2.blur(img2_png, (13,13))

blur_tiff = cv2.blur(img2_tiff, (13,13))
```

### Gaussian Blur

```python
gBlur_jpg = cv2.GaussianBlur (img2_jpg, (13,13), 0)

gBlur_png = cv2.GaussianBlur (img2_png, (13,13), 0)

gBlur_tiff = cv2.GaussianBlur (img2_tiff, (13,13), -1)
```

### median filter

```python
median_jpg = cv2.medianBlur(img2_jpg, 13)

median_png = cv2.medianBlur(img2_png, 13)

median_tiff = cv2.medianBlur(img2_tiff, 13)
```

### Bilateral Filter 

```python
bilateral_jpg = cv2.bilateralFilter(img2_jpg, 13, 75, 75)

bilateral_png = cv2.bilateralFilter(img2_png, 13, 75, 75)

bilateral_tiff = cv2.bilateralFilter(img2_tiff, 13, 75, 75)
```

```python
titles2_jpg = ['Original Image', 'Blur', 'GaussianBlur', 'Median Filter', 'Bilateral Filter']
images2_jpg = [img2_jpg, blur_jpg, gBlur_jpg, median_jpg, bilateral_jpg]
```

```python
for i in range(5):
    plt.subplot(3, 2, i+1), plt.imshow(images2_jpg[i], 'gray')
    plt.title(titles2_jpg[i])
    plt.xticks([]), plt.yticks([])

fig = plt.gcf()
fig.set_size_inches(12.5, 7.5)
```

## Output for PNG file 

```python
titles2_png = ['Original Image', 'Blur', 'GaussianBlur', 'Median Filter', 'Bilateral Filter']
images2_png = [img2_png, blur_png, gBlur_png, median_png, bilateral_png]
```
