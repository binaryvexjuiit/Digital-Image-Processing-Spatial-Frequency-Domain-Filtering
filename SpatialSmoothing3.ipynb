```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

## Load Image

```python
img3_jpg = cv2.imread('e.jpg')
img3_jpg = cv2.cvtColor(img3_jpg, cv2.COLOR_BGR2RGB)

img3_png = cv2.imread('t.png')
img3_png = cv2.cvtColor(img3_png, cv2.COLOR_BGR2RGB)

img3_tiff = cv2.imread('ka.tiff')
img3_tiff = cv2.cvtColor(img3_tiff, cv2.COLOR_BGR2RGB)
```

# Low pass filter 

### Blur 

```python
blur_jpg = cv2.blur(img3_jpg, (13,13)) 

blur_png = cv2.blur(img3_png, (13,13))

blur_tiff = cv2.blur(img3_tiff, (13,13))
```

### Gaussian Blur

```python
gBlur_jpg = cv2.GaussianBlur (img3_jpg, (13,13), 0)

gBlur_png = cv2.GaussianBlur (img3_png, (13,13), 0)

gBlur_tiff = cv2.GaussianBlur (img3_tiff, (13,13), -1)
```

### median filter

```python
median_jpg = cv2.medianBlur(img3_jpg, 13)

median_png = cv2.medianBlur(img3_png, 13)

median_tiff = cv2.medianBlur(img3_tiff, 13)
```

### Bilateral Filter 

```python
bilateral_jpg = cv2.bilateralFilter(img3_jpg, 13, 75, 75)

bilateral_png = cv2.bilateralFilter(img3_png, 13, 75, 75)

bilateral_tiff = cv2.bilateralFilter(img3_tiff, 13, 75, 75)
```

```python
titles3_jpg = ['Original Image', 'Blur', 'GaussianBlur', 'Median Filter', 'Bilateral Filter']
images3_jpg = [img3_jpg, blur_jpg, gBlur_jpg, median_jpg, bilateral_jpg]
```

```python
for i in range(5):
    plt.subplot(3, 2, i+1), plt.imshow(images3_jpg[i], 'gray')
    plt.title(titles3_jpg[i])
    plt.xticks([]), plt.yticks([])

fig = plt.gcf()
fig.set_size_inches(12.5, 7.5)
```

## Output for PNG file 

```python
titles3_png = ['Original Image', 'Blur', 'GaussianBlur', 'Median Filter', 'Bilateral Filter']
images3_png = [img3_png, blur_png, gBlur_png, median_png, bilateral_png]
```
